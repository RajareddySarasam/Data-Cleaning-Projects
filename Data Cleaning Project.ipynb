## Data Cleaning

```python
# Copying the data 

fifa=data.copy()
fifa.sample()
```

```python
# Dealing with the Club column

fifa['Club'].dtype
```

```python
fifa.Club.unique()
```

```python
# TO Remove the newline characters

fifa['Club']=fifa['Club'].str.strip('\n')
fifa['Club'].unique()
```

```python
# Dealing with the Contract column

fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
# We can see that some players are till "on Loan" and few were 'Free'.
# Iam going to create 3 New columns Start_Date,End_Date,Duration
def extract_contract_info(contract):
    if contract=='Free' or 'On Loan' in contract:
        Start_date=np.nan
        End_date=np.nan
        contract_length=0
    else:
        
        Start_date, End_date=contract.split(' ~ ')
        Start_year=int(Start_date[:4])
        End_year=int(End_date[:4])
        contract_length=End_year-Start_year
    return Start_date, End_date,contract_length
        
```

```python
# Applying func to Contract Column 
new_cols=['Start_date', 'End_date','contract_length(years)']
new_data=fifa['Contract'].apply(lambda x : pd.Series(extract_contract_info(x)))

# postioning new cols beside Contract Column
for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i,column=new_cols[i]  , value=new_data[i])
```

```python
fifa.sample(4)
```

```python
# Creating new column Contract Status 

def catergorize_contract_status(contract):
    if contract=='Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else :
        return 'Contract'
# Inserting After contract_length(years)
fifa.insert(loc=fifa.columns.get_loc('contract_length(years)')+1,column='Contract Status',value=fifa['Contract'].apply(lambda x : catergorize_contract_status(x)))
```

```python
fifa.sample(3)
```

```python
# Converting players height into cm

fifa['Height'].unique()
```

```python
# We can that some entrie are in feet and inches

def height_into_cm(height):
    if 'cm' in height:
        return int(height.strip('cm'))
    else:
        feet,inches=height.split("\'")
        total_inches=(int(feet)*12)+int(inches.strip('"'))
        return round(total_inches*2.54)
        
```

```python
height_into_cm('170cm')
```

```python
height_into_cm("6\'4")
```

```python
fifa['Height']=fifa['Height'].apply(height_into_cm)
fifa['Height'].unique()
```

```python
fifa=fifa.rename(columns={'Height':'Height(cm)'})
```

```python
fifa.sample()
```

```python
# Now we are Dealing with  Weight column 

fifa.Weight.dtype
```

```python
fifa['Weight'].unique()
```

```python
# We can see some entries arre in lbs converting those into kg

def weight_into_kg(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        return round(int(weight.strip('lbs'))*0.45359237)
        
```

```python
weight_into_kg('56kg')
```

```python
weight_into_kg('157lbs')
```

```python
fifa['Weight']=fifa['Weight'].apply(weight_into_kg)
fifa['Weight'].unique()
```

```python
fifa=fifa.rename(columns={'Weight':'Weight(kg)'})
```

```python
# we can see that some entries are expressed in k(thousand)
fifa['Hits']=fifa['Hits'].fillna(0)
def Hits_into_int(hits):
    hits=str(hits)
    if 'K' in hits:
        return int(float(hits.strip('K'))*1000)
    elif '.' in hits:
        integer,decimal=hits.split('.')
        return int(integer)
    else:
        return int(hits)
```

```python
Hits_into_int('1.3')
```

```python
type('175')==float
```

```python
fifa['Hits']=fifa['Hits'].apply(Hits_into_int)
fifa['Hits'].unique()
```
