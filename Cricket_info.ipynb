```python
#import libraries 

import pandas as pd 
import numpy as np
```

```python
#Reading Csv file
data=pd.read_csv('cricinfo.csv')
```

```python
data.head()
```

```python
data=data.rename(columns={'Mat':'Matches','Inns':'Innings','NO':'Notouts','HS':'Highest score','BF':'Balls Faced','SR':'Strike Rate'})
data.head()
```

```python
#Checking for Null values 

data.isnull().sum()
```

```python
# Checking for Duplicates 

data[data.duplicated()==1]
```

```python
# We can see that three entries are duplicated 

data[data['Player'].isin(['GA Headley (WI)','GS Sobers (WI)','KS Williamson (NZ)'])]
```

```python
data=data.drop_duplicates()
```

```python
data[data['Player'].isin(['GA Headley (WI)','GS Sobers (WI)','KS Williamson (NZ)'])]
```

```python
# So,We Have Dropped Duplicates

# let's Split Span Column into 2 columns of start ,final year
First_year=data['Span'].str.split(pat='-').str[0]
```

```python
Final_year=data['Span'].str.split(pat='-').str[1]
```

```python
data.insert(loc=1,column='First year',value=First_year)
```

```python
data.insert(loc=2,column='Final year',value=Final_year)
```

```python
data.head()
```

```python
data=data.drop('Span',axis=1)
```

```python
# Will Continue rest of prject  tommorow
```

```python
# Creating a country column


data['Country']=data['Player'].str.split(pat='(').str[1]
```

```python
data['Country']=data['Country'].str.split(pat=')').str[0]
```

```python
data['Country']
```

```python
# Removing the Country from Player Name 

data['Player']=data["Player"].str.split(pat='(').str[0]
```

```python
data['Player']
```

```python
# Checking the datatypes of all cloumns 

data.dtypes
```

```python
data['First year']=data['First year'].astype('int64')
```

```python
data['Final year']=data['Final year'].astype('int64')
```

```python
data.head()
```

```python
def HS(score):
    if '*' in score:
        return int(score[:-1])
    return int(score)
```

```python
HS('125*')
```

```python
data['Highest score']=data['Highest score'].apply(HS)
```

```python
# Droping a entry and reseting index
data=data.drop(55,axis=0)
data.reset_index()
```

```python
# Creating a col for CAreer LEngth

data['Career Length']=data['Final year']-data['First year']
```

```python
#Q1 -> Avg career length
data['Career Length'].mean()
```

```python
#Q2 -> Avg Matches Played of player who played mmore than 10 years

int(data[data['Career Length'] > 10]['Matches'].mean())
```
